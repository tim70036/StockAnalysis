```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

r = np.array(range(9000, 11000)) / 10000
p = (( (2**(1/3) + (0.25)**(1/3)) * (r**(2/3)) ) / ( (2**(1/3))*r + (0.25**(1/3)) )     ) - 1
plt.xlabel('r')
plt.ylabel('P')
plt.plot(r, p)
```

```python
r = np.array(range(9000, 11000)) / 10000
p = ( (2 * (r**(1/2))) / (r+1)  ) - 1
plt.xlabel('r')
plt.ylabel('P')
plt.plot(r, p)
```

```python
p = 0.5
u = np.array(range(0, 4000001))
mean = p * (3000000 - u) + (1-p) * (u - 1000000)
var = p * ((3000000 - u - mean)**2) + (1-p) * ((u - 1000000 - mean)**2)
std = var ** (1/2)
plt.plot(mean, std)

p = 0.6
u = np.array(range(0, 4000001))
mean = p * (3000000 - u) + (1-p) * (u - 1000000)
var = p * ((3000000 - u - mean)**2) + (1-p) * ((u - 1000000 - mean)**2)
std = var ** (1/2)

plt.title('std vs. mean as u change from 0 to 4000000')
plt.xlabel('mean')
plt.ylabel('std')
plt.plot(mean, std)



print (mean)
```

```python
r_free = 0.04
r_risky = 0.12
std_risky = 0.2

std_p = np.array(range(0, 300)) / 1000
r_p = std_p * ((r_risky - r_free) / std_risky) + r_free

plt.title('std vs. epected return')
plt.xlabel('standard deviation of portfolio')
plt.ylabel('return of portfolio')
plt.plot(std_p, r_p)
```

```python
r_free = 0.04
r_free_change = 0.06

r_risky = 0.12
std_risky = 0.2

std_p1 = np.array(range(0, 201)) / 1000
r_p1 = std_p1 * ((r_risky - r_free) / std_risky) + r_free

std_p2 = np.array(range(201, 400)) / 1000
r_p2 = std_p2 * ((r_risky - r_free_change) / std_risky) + r_free_change




plt.title('std vs. epected return')
plt.xlabel('standard deviation of portfolio')
plt.ylabel('return of portfolio')
plt.plot(np.append(std_p1, std_p2), np.append(r_p1, r_p2))
```
